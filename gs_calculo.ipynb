```python

anos = ['2014', '2015', '2016']
agendadas = [7365, 6317, 10013]
confirmadas = [4803, 3724, 5598]
abstencao = [2566, 2593, 4415]
valores_desperdicados = [52159.60, 29000.00, 45470.00]

# Calcular a porcentagem de abstenção
porcentagem_abstencao = [round((a / c) * 100, 2) for a, c in zip(abstencao, confirmadas)]

# Gráfico de barras para consultas agendadas, confirmadas e abstenção
plt.figure(figsize=(12, 6))
plt.bar(anos, agendadas, label='Agendadas', color='lightblue', alpha=0.7)
plt.bar(anos, confirmadas, label='Confirmadas', color='blue', alpha=0.7)
plt.bar(anos, abstencao, label='Abstenção', color='red', alpha=0.7)
plt.xlabel('Ano')
plt.ylabel('Número de Consultas')
plt.title('Consultas Agendadas, Confirmadas e Abstenção por Ano (medico cardiologista)ES')
plt.legend()
plt.show()

# Gráfico de pizza para a porcentagem de abstenção
plt.figure(figsize=(12, 6))
plt.pie(porcentagem_abstencao, labels=anos, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green'])
plt.title('Porcentagem de Abstenção por Ano (medico cardiologista 2014-2016 )ES')
plt.show()
```

```python

dados = {
    '2014': {'agendadas': 7365, 'confirmadas': 4803, 'abstencao': 2566, 'prejuizo': 52159.60},
    '2015': {'agendadas': 6317, 'confirmadas': 3724, 'abstencao': 2593, 'prejuizo': 29000.00},
    '2016': {'agendadas': 10013, 'confirmadas': 5598, 'abstencao': 4415, 'prejuizo': 45470.00}
}

# Calcular abstenção em porcentagem e adicionar ao dicionário
for ano, valores in dados.items():
    valores['abstencao_pct'] = (valores['abstencao'] / valores['confirmadas']) * 100

# Calcular valor médio de prejuízo
prejuizo_medio = sum(valores['prejuizo'] for valores in dados.values()) / len(dados)

# Imprimir resultados

print(f"Valor médio de prejuízo: R$ {prejuizo_medio:.2f}")
```
